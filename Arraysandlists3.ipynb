### Arrays and Lists
In this notebook, I want to use graphs to show the amount of money  
that every person has (after iterations and it is related to the  
program that I wrote in the last notbook named Arraysandlists2.ipynb).

```python
people=[100 for _ in range(50)]
```

```python
import random
iterations=10**6
i=0
while i<iterations and len(people)>=2:
    i+=1
    random_person1,random_person2=random.sample(range(len(people)),2)
    if people[random_person2]>0:
        people[random_person1]+=1
        people[random_person2]-=1
    people=[person for person in people if person>0]
    if len(people)<2:
        print("There isn't enought pepole for the task.")
        break
people.sort(reverse=True)
```

In the cell below I tried to learn how matplotlib actually works  
and I made a simple bar chart.

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('seaborn-v0_8-whitegrid')

x = 0.5 + np.arange(10)
y = [4.8 , 8.0 , 5.4 , 4.7 , 1.9 , 2.9 , 5.6 , 7.4 , 0.0 , 0.0]
fig, ax = plt.subplots()
ax.bar(x, y ,width=1, edgecolor="white" , linewidth=0.7)
ax.set(xlim=(0, 10), xticks=np.arange(1,10) , ylim=(0,10) , yticks=np.arange(1,10) )
plt.show()
```

In the next cell I will try to make a bar chart that shows  
how much money every person has (fter the iterations of the  
money that was exchanged between them). 

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-v0_8-whitegrid')
x = np.arange(len(people))
fig, ax = plt.subplots(figsize=(10,5))
ax.bar(x, people ,width=0.3, edgecolor="white" , linewidth=0.7)
ax.set_title("People's money", fontsize=16)
ax.set_xlabel("People", fontsize=14)
ax.set_ylabel("Money", fontsize=14)
ax.set(xlim=(0, 49), xticks=np.arange(0,50,5) , ylim=(0,1000) , yticks=np.arange(0,1000,50) )
plt.show()
```
