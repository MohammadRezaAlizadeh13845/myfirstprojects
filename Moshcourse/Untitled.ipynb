```python
import matplotlib.pyplot as plt
import time

def emojis():
    message=input(">")
    contetnt=message.split(' ')
    emojis={":D":"😁",":(":"😒",":|":"😐"}
    output=""
    for item in contetnt:
        output+= emojis.get(item,item)+" "
    print(output)


class Plotter:
    def __init__(self):
        self.fig, self.ax = plt.subplots()
        self.ax.set_xlim(0, 10)
        self.ax.set_ylim(0, 10)
        self.ax.set_xlabel("X-axis")
        self.ax.set_ylabel("Y-axis")
        self.ax.set_title("Dynamic Point Plotter")

    def plot_points(self,x,y):
        self.ax.scatter(x, y, color='red', marker='o', s=100)
        plt.draw()
        plt.show() 


nums = input("Enter the coordinates of the points : ")

xs = [int(nums[i]) for i in range(0, len(nums), 2)]
ys = [int(nums[i+1]) for i in range(0, len(nums), 2)]

plotter = Plotter()
plotter.plot_points(xs,ys)
```
